# 서울 구별로 나눈 지도 로딩

```python
import json
import folium
import numpy as np
import pandas as pd

geo_path ='C:/Users/hello/Documents/DataScience/data/02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
gu_names=['강동구','강남구','강북구','강서구','관악구','광진구','구로구','금천구','노원구','도봉구',
  '동대문구','동작구','마포구','서대문구','서초구','성동구','성북구','송파구','양천구','영등포구',
  '용산구','은평구','종로구','중구','중랑구']
```

# 후보 1 ~ 후보 4의 가상 투표울 만들기 (후보 1 ~ 후보4를 숫자(int)로 구별함)

```python
no1=pd.Series(np.random.randn(25),index=gu_names,name=-20)
no2=pd.Series(np.random.randn(25),index=gu_names,name=-10)
no3=pd.Series(np.random.randn(25),index=gu_names,name=0)
no4=pd.Series(np.random.randn(25),index=gu_names,name=10)
A=no1
B=no2
C=no3
D=no4
```

```python
result=pd.concat([A,B,C,D], axis=1)
result
```

```python
maxdf=result.idxmax(axis=1)
result['최고 득표자']=maxdf
result
```

```python
voting_result=result.loc[:,['최고 득표자']] #구 & 최고 득표자로만 table 구성함
voting_result
```

# 구별 최고 득표자 후보 색 나타내기 (추후 업데이트 필요)

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11, 
                 tiles='Stamen Terrain')

map.choropleth(geo_data = geo_str,
               data = voting_result['최고 득표자'],
               columns = [voting_result.index, voting_result['최고 득표자']],
               fill_color = 'PuRd', #PuRd, YlGnBu
               key_on = 'feature.id')
map
```

# selenium으로 해당 사진 스크린샷 찍기

```python
import os
import time
from selenium import webdriver

delay=5
fn='testmap.html'
tmpurl='file://{path}/{mapfile}'.format(path=os.getcwd(),mapfile=fn)
map.save(fn)

browser = webdriver.Chrome('C:\chromedriver_win32/chromedriver.exe')
browser.get(tmpurl)
#Give the map tiles some time to load
time.sleep(delay)
browser.save_screenshot('map.png')
browser.quit()
```
